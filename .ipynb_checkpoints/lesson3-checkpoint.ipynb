### замена пустых на медианы 

```python
df.loc[df['housing_median_age'].isnull(),'housing_median_age'] = df['housing_median_age'].median()
```

```python
df.loc[df['total_bedrooms'].isnull(),'total_bedrooms'] = df['total_bedrooms'].median()
```

```python
df.loc[df['population'].isnull(),'population'] = df['population'].median()
```

```python
df.describe()
```

```python
df['ocean_proximity'].mode()[0]
```

```python
df.replace({'ocean_proximity' : {'-': df['ocean_proximity'].mode()[0]}}, inplace =True)
```

```python
df['ocean_proximity'].value_counts()
```

```python
df.describe()
```

## правка выбросов

```python
df[df['total_rooms'] > 20000]
```

```python
df.loc[df['longitude'] > 0,'longitude'] = df.loc[df['longitude'] >=0,'longitude'] * -1
```

```python
df.loc[df['longitude'] == 0,'longitude'] = df['longitude'].median()
```

```python
df[df['longitude'] >=0]
```

```python
df.loc[df['latitude'] >=50, 'latitude'] = df['latitude'].median()
```

```python
df.loc[df['latitude'] <=0, 'latitude'] = df['latitude'].median()
```

## Исключаем признак ID

```python
df = df[df.columns[:-1]]
```

## перобразуем ocean_proximity в несколько бинарых

```python
df = pd.concat([df, pd.get_dummies(df['ocean_proximity'])], axis = 1)
```

```python
df.head()
```

```python
df.to_csv(DATASET_PREP, index = False, encoding = 'UTF-8')
```

### Загрузка данных

#### Описание датасета

Статистические данные о ряде домов в Калифорнии, основанные на переписи 1990 года.

longitude - долгота  
latitude - широта  
housing_median_age - средний возраст дома  
total_rooms - общее количество комнат 
total_bedrooms - общее количество спален  
population - количество проживающих 
households - домохозяйства 
ocean_proximity - близость океана  
median_income - средний доход  
median_house_value - средняя стоимость дома  

```python
df = pd.read_csv(DATASET_PREP)
df.head()
```

```python
df.dtypes
```

### Анализ целевой переменной

```python
plt.figure(figsize= (17, 8))

plt.subplot(121)
df['median_house_value'].hist()

plt.ylabel('count')
plt.xlabel('median_house_value')

plt.subplot(122)

sns.kdeplot(df['median_house_value'], shade=True, legend=False)
plt.xlabel('median_house_value')

plt.suptitle('Distribution of median_house_value')
plt.show()
```

```python
target_mean = round(df['median_house_value'].mean(), 2)
target_median = df['median_house_value'].median()
target_mode = df['median_house_value'].mode()[0]

target_mean, target_median, target_mode
```

```python
plt.figure(figsize = (16, 8))

sns.distplot(df['median_house_value'])


y = np.linspace(0, 0.000005, 100)

plt.plot([target_mean] * 100, y, label='mean', linestyle=':', linewidth=4)

plt.plot([target_median] * 100, y, label='median', linestyle='--', linewidth=4)

plt.plot([target_mode] * 100, y, label='mode', linestyle='-.', linewidth=4)

plt.title('Distribution of median_house_value')
plt.legend()
plt.show()
```

### Оценка распредения целевой переменной в разрезе других признаков

```python
plt.figure(figsize=(16, 8))

plt.scatter(df['median_house_value'], df['ocean_proximity'])

plt.xlabel('median_house_value')
plt.ylabel('ocean_proximity')
plt.title('Distribution of median_house_value by ocean_proximity')
plt.show()
```

### Box plot (Ящик с усами)

```python
plt.figure(figsize=(16, 8))

sns.boxplot(df['median_house_value'], df['ocean_proximity'])

plt.xlabel('median_house_value')
plt.ylabel('ocean_proximity')
plt.title('Distribution of median_house_value by ocean_proximity')
plt.show()
```

### Violin plot

```python
plt.figure(figsize=(16, 8))

sns.violinplot(df['median_house_value'], df['ocean_proximity'])

plt.xlabel('median_house_value')
plt.ylabel('ocean_proximity')
plt.title('Distribution of median_house_value by ocean_proximity')
plt.show()
```
