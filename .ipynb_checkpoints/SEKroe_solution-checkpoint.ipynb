### Импорт библиотек

```python
import numpy as np
import pandas as pd
import pickle

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import r2_score as r2, mean_absolute_error as mae, mean_squared_error as mse
from sklearn.model_selection import KFold, GridSearchCV

import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
def evaluate_preds(true_values, pred_values):
    print("R2:\t" + str(round(r2(true_values, pred_values), 3)) + "\n" +
          "MAE:\t" + str(round(mae(true_values, pred_values), 3)) + "\n" +
          "MSE:\t" + str(round(mse(true_values, pred_values), 3)))
    
    plt.figure(figsize=(10,10))
    
    sns.scatterplot(x=pred_values, y=true_values)
    
    plt.xlabel('Predicted values')
    plt.ylabel('True values')
    plt.title('True vs Predicted values')
    plt.show()
```

### Обработка исходных данных

#### Исключаем признак ID

```python
frame = frame[frame.columns[1:]]
```

#### Поиск пустых значений

```python
frame.info()
```

#### замена пустых и экстремальных признаков Healthcare_1 на медианное значение

```python
frame.loc[frame['Healthcare_1'].isnull(),'Healthcare_1'] = frame['Healthcare_1'].median()
```

```python
frame.loc[frame['Healthcare_1'] < 30 ,'Healthcare_1'] = frame['Healthcare_1'].median()
```

```python
frame.loc[frame['Rooms'] > 6, 'Rooms'] = frame['Rooms'].median()
```

```python
frame.loc[frame['Square'] >= 500, 'Square' ] = frame['Square'].median()
```

```python
frame.loc[frame['Rooms'] == 0, 'Rooms'] = frame['Rooms'].median()
```

```python
frame.loc[frame['KitchenSquare'] > frame['Square'],'KitchenSquare'] = frame['KitchenSquare'].median()
```

```python
frame.describe()
```

##### вычисление отношения жилой прощади к нежилой у нормальных прощадей

```python
df = frame[frame['LifeSquare'].notnull() & (frame['Square'] > frame['LifeSquare'])]
```

```python
df['l/s'] = df['Square']/ df['LifeSquare']
```

#### проверка где общая площадь меньше жилой

```python
frame[frame['Square'] <= frame['LifeSquare']]
```

```python
df['l/s'].median()
```

#### замена площадей. 

```python
frame.loc[frame['Square'] <= frame['LifeSquare'], 'Square'] = df['l/s'].median() *  frame['LifeSquare']
```

```python
frame.loc[frame['LifeSquare'].isnull(), 'LifeSquare'] = frame['Square']/ df['l/s'].median()
```
